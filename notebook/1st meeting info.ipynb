# Open hardware researchers group
### 1st survey results

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv (r'survey.csv')
```

```python
where = sns.catplot(x="TimeZone", kind="count", palette="ch:.25", data=df, order = df['TimeZone'].value_counts().index)
plt.xticks(rotation = 45)
plt.title("Where in the world are we?");
```

```python
sns.catplot(x="Platform", kind="count", palette="ch:.25", data=df)
plt.title("Preferred platform for meeting");
```

```python
sns.catplot(x="Where do you develop most of your work related to this group?", kind="count", palette="ch:.25", data=df).set(xlabel=None)
plt.title("Where do you develop most of your work related to this group?");
```

```python
df2 = df[["I'm a researcher looking for collaborations", "I want to increase the visibility/impact of my research", "I want to be part of a community of people working on similar stuff", "I want to be on top of what's going on in the field", "I want to learn about open hardware"]]
df2 = pd.melt(df2)
```

```python

df2.dropna(subset = ["value"], inplace=True)
```

```python

why = sns.catplot(y="variable", kind="count", palette="ch:.25", data=df2, orient="v", order = df2['variable'].value_counts().index).set(ylabel=None)
plt.xticks(rotation = 90)
plt.title("Why do you want to join the group?");

```

```python
df3 = df[["Time 1", "Time 2", "Time 3"]]
df3 = pd.melt(df3)

df3.dropna(subset = ["value"], inplace=True)
sns.catplot(y="variable", kind="count", palette="ch:.25", data=df3).set(ylabel=None)
plt.title("Which is the best time for you to meet?");
```
